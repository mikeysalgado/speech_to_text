```python
import os
import pandas as pd
import re
import csv
import librosa
from pydub import AudioSegment
```

```python
transcripts_path = 'C:/Users/salgadom/Documents/CS 4980 003/Data/transcript/'
audio_path = 'C:/Users/salgadom/Documents/CS 4980 003/Data/output_files/'
wav_path = 'C:/Users/salgadom/Documents/CS 4980 003/Data/wav_files'
flac_path = 'C:/Users/salgadom/Documents/CS 4980 003/Data/flac_files'
```

```python
def convert_mp3_to_wav(a_path, wav_path):
    if not os.path.exists(wav_path):
        os.makedirs(wav_path)

    audio_data = []
    for ted_talk_folder in os.listdir(a_path):
        folder_path = os.path.join(a_path, ted_talk_folder)
        if os.path.isdir(folder_path):
            folder_name = ted_talk_folder.replace(" ", "_")  # replace spaces with underscores
            wav_folder_path = os.path.join(wav_path, folder_name)
            os.makedirs(wav_folder_path, exist_ok=True)  # create folder if it doesn't exist
            mp3_files = [f for f in os.listdir(folder_path) if f.endswith('.mp3')]
            sorted_files = sorted(mp3_files, key=lambda x: int(x.split('_')[-1].split('.')[0]))
            for file in sorted_files:
                mp3_path = os.path.join(folder_path, file).replace('\\', '/')
                if os.path.isfile(mp3_path):
                    wav_file = os.path.splitext(file)[0] + ".wav"
                    wav_file_path = os.path.join(wav_folder_path, wav_file)
                    if not os.path.exists(wav_file_path):
                        sound = AudioSegment.from_mp3(mp3_path)
                        print(f'Exporting {wav_file_path}...')
                        sound.export(wav_file_path, format="wav")
                    audio_data.append(wav_file_path)
    return audio_data
```

```python
def convert_mp3_to_flac(a_path, flac_path):
    if not os.path.exists(flac_path):
        os.makedirs(flac_path)

    audio_data = []
    for ted_talk_folder in os.listdir(a_path):
        folder_path = os.path.join(a_path, ted_talk_folder)
        if os.path.isdir(folder_path):
            folder_name = ted_talk_folder.replace(" ", "_")  # replace spaces with underscores
            flac_folder_path = os.path.join(flac_path, folder_name)
            os.makedirs(flac_folder_path, exist_ok=True)  # create folder if it doesn't exist
            mp3_files = [f for f in os.listdir(folder_path) if f.endswith('.mp3')]
            sorted_files = sorted(mp3_files, key=lambda x: int(x.split('_')[-1].split('.')[0]))
            for file in sorted_files:
                mp3_path = os.path.join(folder_path, file).replace('\\', '/')
                if os.path.isfile(mp3_path):
                    flac_file = os.path.splitext(file)[0] + ".flac"
                    flac_file_path = os.path.join(flac_folder_path, flac_file)
                    if not os.path.exists(flac_file_path):
                        sound = AudioSegment.from_mp3(mp3_path)
                        print(f'Exporting {flac_file_path}...')
                        sound.export(flac_file_path, format="flac")
                    audio_data.append(flac_file_path)
    return audio_data
```
